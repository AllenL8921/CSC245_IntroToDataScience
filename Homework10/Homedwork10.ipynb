Create a basic line plot of the function y = sin(x) over the interval [0, 2*pi]. Label the axes and add a title.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of y = sin(x)')
plt.grid(True)
plt.show()
```

Section 4.3 â€“ Multiple Plots and Subplots


Use subplot() to create a 2x2 grid of plots for sine, cosine, tangent, and exponential functions.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
functions = [np.sin, np.cos, np.tan, lambda x: np.exp(-x)]
titles = ['sin(x)', 'cos(x)', 'tan(x)', 'exp(-x)']

plt.figure(figsize=(10, 8))

for i, func in enumerate(functions, 1):
    plt.subplot(2, 2, i)
    plt.plot(x, func(x))
    plt.title(titles[i-1])
    if titles[i-1] == 'tan(x)':
        plt.ylim(-5, 5)  # limit for tan(x)

plt.tight_layout()
plt.show()
```


Compare subplot() vs tiledlayout(). Plot the same functions using both methods.


```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

funcs = [np.sin, np.cos, np.tan, lambda x: np.exp(-x)]
titles = ['sin(x)', 'cos(x)', 'tan(x)', 'exp(-x)']

for ax, func, title in zip(axs.flat, funcs, titles):
    ax.plot(x, func(x))
    ax.set_title(title)
    if title == 'tan(x)':
        ax.set_ylim(-5, 5)

plt.tight_layout()
plt.show()
```


Use linkaxes() to synchronize zooming/panning across multiple plots.


```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create first subplot
ax1 = plt.subplot(2, 1, 1)   # 2 rows, 1 column, 1st subplot
ax1.plot(x, y1)
ax1.set_title('sin(x)')

# Create second subplot, sharing the x-axis with ax1
ax2 = plt.subplot(2, 1, 2, sharex=ax1)
ax2.plot(x, y2)
ax2.set_title('cos(x)')

# Add axis labels
ax1.set_ylabel('Amplitude')
ax2.set_xlabel('x')
ax2.set_ylabel('Amplitude')

plt.setp(ax1.get_xticklabels(), visible=False)

plt.tight_layout()
plt.show()
```

Section 4.5 â€“ 3D Plotting


Plot a 3D surface using surf() for z = sin(x) * cos(y).

Add colorbars and shading to improve visualization.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Needed for 3D plots

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
plt.title('3D Surface: sin(x) * cos(y)')
plt.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```


Use meshgrid() and mesh() to plot wireframe surfaces.


```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, color='gray', linewidth=0.5)
plt.title('3D Wireframe Plot')
plt.show()
```


Animate rotation of a 3D plot using view() in a loop.


```python
import matplotlib.animation as animation

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='plasma')

def rotate(angle):
    ax.view_init(elev=30, azim=angle)

ani = animation.FuncAnimation(fig, rotate, frames=np.arange(0, 360, 2), interval=100)
plt.show()
```


Use contour3() to add contours to a 3D surface.



```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='inferno')
ax.set_title('3D Contour Plot')
plt.show()
```

Section 4.6 â€“ Interactive and Advanced Plotting


Use ginput() to capture mouse clicks and annotate those points.


```python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Click on the plot to capture points")

# Capture 3 points
points = plt.ginput(n=3, timeout=0)

# Annotate the clicked points
for point in points:
    plt.annotate(f'({point[0]:.2f}, {point[1]:.2f})', 
                 (point[0], point[1]), textcoords="offset points", 
                 xytext=(0,10), ha='center', color='red')

plt.show()
```


Add interactivity using uicontrol() elements like sliders.


```python
from matplotlib.widgets import Slider
import numpy as np
import matplotlib.pyplot as plt

# Initial plot setup
x = np.linspace(0, 10, 100)
y = np.sin(x)
fig, ax = plt.subplots()
line, = ax.plot(x, y)

# Slider axis
ax_slider = plt.axes([0.2, 0.01, 0.65, 0.03], facecolor='lightgoldenrodyellow')
slider = Slider(ax_slider, 'Frequency', 0.1, 10.0, valinit=1)

# Update function for slider
def update(val):
    freq = slider.val
    line.set_ydata(np.sin(freq * x))
    fig.canvas.draw_idle()

slider.on_changed(update)
plt.show()
```


Create an animated line plot using animatedline().


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots()
line, = ax.plot([], [], lw=2)
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1.1, 1.1)

# Initialization function
def init():
    line.set_data([], [])
    return line,

# Update function for animation
def update(frame):
    line.set_data(x[:frame], y[:frame])
    return line,

ani = FuncAnimation(fig, update, frames=range(1, len(x)+1), init_func=init, blit=True, interval=50)

plt.show()
```


Combine plot3() and comet3() to show 3D trajectories over time.



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# Generate 3D data (spiral trajectory)
theta = np.linspace(0, 2*np.pi, 100)
z = np.linspace(0, 10, 100)
x = np.sin(theta) * z
y = np.cos(theta) * z

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([0, 10])

# Initial plot setup
line, = ax.plot([], [], [], marker='o', markersize=5)

# Initialization function
def init():
    line.set_data([], [])
    line.set_3d_properties([], [])
    return line,

# Update function for animation
def update(frame):
    line.set_data(x[:frame], y[:frame])
    line.set_3d_properties(z[:frame])
    return line,

ani = FuncAnimation(fig, update, frames=len(x), init_func=init, blit=True, interval=100)

plt.show()
```
